## Setting up our Plotly

```python
import chart_studio.plotly as px #chartstudio is good lib for charts
import plotly.express as exp #express has different functionality
import plotly.graph_objects as go #graphobjects has more function than express but both have
#different use case
#Main difeerence is Graph_objects has more customization option than Express
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
#plot and iplot are used to show your graphs or data viz in jupyter notebook
init_notebook_mode(connected=True)
#this above line help us show our graphs in Jupyter notebook and not in browser as Jupyter 
#notebook is hosted on top of a browser
```

## Plotting Line, Bar, Pie, Scatter and Histograms

```python
days = ['mon','tue','weds','thurs']
sale = [10000,20000,5000,50000]
exp.line(x = days, y = sale)
```

```python
#lets plot a bar chart
import pandas as pd
df = pd.read_csv(r'/Users/csuftitan/Downloads/Matplotlib/zomato_sample.csv')
```

```python
barGraph = go.Bar(x= df['book_table'].value_counts().index, 
                                              y = df['book_table'].value_counts().values)
```

```python
iplot([barGraph]) # need iplot to see the graph
```

```python
#plotting the pie chart for this
exp.pie(df, values=df['book_table'].value_counts().values, names=df['book_table'].value_counts().index)
```

```python
#plotting Histogram, so whenever you use graph object you need iplot to show it
hist = go.Histogram(x = df['rate'])
iplot([hist])
```

```python
#plotting scatter plot using express
exp.scatter(x = df['rate'], y =df['approx_cost(for two people)'])
```

## Changing colors of our Plots

```python
exp.scatter(x = df['rate'], y =df['approx_cost(for two people)'], color=df['rate'])
```

```python
exp.scatter(x = df['rate'], y =df['approx_cost(for two people)'], color=df['approx_cost(for two people)'])
```

```python
#lets change the default template to dark
exp.scatter(x = df['rate'], y =df['approx_cost(for two people)'], template='plotly_dark', color=df['rate'])
```

### Changing Height and Width of Plot

```python
# the height & width concerns with the pixels we utlizie in our plot
exp.scatter(x = df['rate'], 
            y =df['approx_cost(for two people)'], 
            color=df['approx_cost(for two people)'],
           height=800,
           width=800)
```

## Sharing our Plotly dashboard using Dash(based on Flask)

```python
barGraph = go.Bar(x= df['book_table'].value_counts().index, 
                                              y = df['book_table'].value_counts().values)

#need to use it as figure to display in dash
fig = go.Figure(data = barGraph)
```

```python
#to use Dash we need its HTML component as well its Core Component
# In general we need Dash, HTML comp and Core Comp
!pip install dash
```

```python
import dash
import dash_html_components as html
import dash_core_components as core
```

```python
app = dash.Dash()
```

```python
# now we need to provide layout of our HTML
app.layout = html.Div([core.Graph(figure=fig)])
```

```python
app.run_server(debug=True, use_reloader = False)
```
