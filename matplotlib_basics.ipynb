```python
##Using bar graph to check how people like to book table from our amazon data
df.head(10)
```

```python
df['book_table'].value_counts().values # gives the count of two values
```

```python
df['book_table'].value_counts().index # gives index of those value
```

```python
#now lets plot it using a bar graph
plt.bar( x = df['book_table'].value_counts().index,
       height=df['book_table'].value_counts().values)
```

```python
#plotting pie chart for the same where index 'Yes/No' would be used as values 
# the values would used to determine pie size
plt.pie(labels = df['book_table'].value_counts().index,
       x = df['book_table'].value_counts().values)
```
