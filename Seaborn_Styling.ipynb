## Adding Grid to our Scatterplot and changing colors

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'/Users/csuftitan/Downloads/Matplotlib/zomato_sample.csv')
df.head()
```

```python
sns.set_style('darkgrid') #use shift+tab inside function to check documentation
sns.scatterplot(x = df['rate'], y = df['approx_cost(for two people)'], color = "red")
#red colour scatterplot with grids
```

## Chaning bin size and title for a histogram

```python
axs = sns.histplot(df['rate'], bins=30) #by default bin size is 10
axs.set_title("Histogram")
```

## Changing Figure Size

```python
sns.set(rc= {'figure.figsize': (10,10)})
sns.countplot(df['rest_type']) #here the x-axis is messed up lets make it good in next change
```

## Rotate our X Labels

```python
sns.set(rc= {'figure.figsize': (10,10)})
az = sns.countplot(df['rest_type'])
az.set_xticklabels(az.get_xticklabels(), rotation=90)
#this above method gives lot of additional unnecessary data
#so to get rid of it we import plt.show from matplotlib
#Seaborn is based on matplot lib
import matplotlib.pyplot as plt
plt.show()
```

## Setting the X axis and Y axis Limits

```python
sns.set_style('darkgrid') #use shift+tab inside function to check documentation
acc = sns.scatterplot(x = df['rate'], y = df['approx_cost(for two people)'], color = "red")
acc.set(xlim=(4.50,5.00), ylim=(1500,3000))
```

```python
#another way setting the limits
acc = sns.scatterplot(x = df['rate'], y = df['approx_cost(for two people)'], color = "red")
acc.set(xlim=(3.50,5.00))
acc.set(ylim=(1000,3000))
```

### Hue in seaborn - mostly utilized for cluserting where we have multiple data for example in a scatterplot

```python
acc = sns.scatterplot(x = df['rate'], y = df['approx_cost(for two people)'], hue=df['book_table'])
#so this means people who have booked table based upon good rating and medium budget
```
