# Sorting Data Values in Pandas

```python
#using sort_values function
df.sort_values(by = 'Score', ascending= False) #sort based upon score in Descending order
```

# Using Custom Functions(User Defined) on Series and dataframes

```python
#lets create a function that counts the lenght of "Text" column and then append this length to dataframe
def count_length(x):
    return len(x)


```

```python
df["Text_Length"] = df['Text'].apply(count_length)
```

```python
df.head(4)
```

```python
#Using Lambda function 
df['Text_Length'].apply(lambda x : x+10)
```
