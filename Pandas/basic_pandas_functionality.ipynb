## Missing Values in Data and How to deal with them

1. Deletion - deleting all missing values. This may cause missing out on values.

2. Imputation - just to impute that missing value or impute  using some statistical technique.

### finding missing values using isnull or isna function

So After finding those missing values we can either drop them dropna (deletion) or fill them fillna (Imputation)

```python
df.isnull() #used in earlier pandas version
```

```python
df.isna() #used in recent pandas version
```

```python
df.isnull().sum() #gives total Null Values per column
```

```python
#If you have few missing values then its good to delete them
df.dropna(subset=['HelpfulnessNumerator','Score','HelpfulnessDenominator'])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.fillna(0,inplace=True) #Inplace help us update the data frame itself thts why didnt added in dropna function
```

```python
df.isnull().sum()
```

```python
df.head()
```
