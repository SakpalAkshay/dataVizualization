### Customizing the Size of the plot

```python
#let us first see the bar graph of best restaurents in our sample file
df['rest_type']
```

```python
df['rest_type'].value_counts()
```

```python
#lets plot the bar graph using values and indexes
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values) #this does give out any info 
```

```python
#let try zooming in on our bar graph
plt.figure(figsize=(20,12))
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values) #still can see the x labels
```

```python
#first way of sorting the x labels
#plt.gcf()==> gives current figure (bar graph) .autofmt_xdate() ==> formats the x label
plt.figure(figsize=(20,12))
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values)
plt.gcf().autofmt_xdate() 
```

```python
#second way of doing it
plt.figure(figsize=(20,12))
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values)
plt.xticks(rotation = 'vertical') #this gives lot of info, to get just the grap use plt.show()
plt.show()
```

## Setting Limits on your axis

```python
#lets say you have a scatter plot and you want to get rating data which are good only
# i.e above 3.0 only
#lets see out basic scatter plot with a bit resizing
plt.figure(figsize=(7,7))
plt.scatter(x=df['rate'], y = df['approx_cost(for two people)'])
plt.title("Scatterplot")
plt.xlabel("Rating for food")
plt.ylabel("Approx Cost of Food")
```

```python
#lets set our limits
plt.figure(figsize=(7,7))
plt.scatter(x=df['rate'], y = df['approx_cost(for two people)'])
plt.title("Scatterplot")
plt.xlabel("Rating for food")
plt.ylabel("Approx Cost of Food")
plt.xlim(2.99,5.00) # limitting x for ratings above 2.99
plt.ylim(500,2000) #finding restaurents between budget of 500 and 2000
```

```python
#lets find the most expensive restaurent with 4.5 above rating
plt.figure(figsize=(7,7))
plt.scatter(x=df['rate'], y = df['approx_cost(for two people)'])
plt.title("Scatterplot")
plt.xlabel("Rating for food")
plt.ylabel("Approx Cost of Food")
plt.xlim(4.50,5.00) # limitting x for ratings above 2.99
plt.ylim(2000,3000)
```
